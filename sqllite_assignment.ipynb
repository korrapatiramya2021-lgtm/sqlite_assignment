```python
from sqlite3.dbapi2 import connect
import sqlite3
conn=sqlite3.connect("bookstore.db")
Cursor=conn.cursor()
#Book table
Cursor.execute("""CREATE TABLE IF NOT EXISTS books(book_id INTEGER PRIMARY KEY AUTOINCREMENT,
title TEXT Not Null,author TEXT Not Null,price REAL Not Null,
stock_quantity INTEGER Default 0
)""")
```

```python
#customers table
Cursor.execute("""CREATE  TABLE IF NOT EXISTS customers(customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT Not Null,
email TEXT Unique Not Null,
city TEXT,
join_date TEXT)
""")
```

```python
#orders table
Cursor.execute("""CREATE TABLE IF NOT EXISTS orders(
order_id INTEGER PRIMARY KEY AUTOINCREMENT,
customer_id INTEGER,
book_id INTEGER,
quantity INTEGER NOT NULL,
order_date TEXT NOT NULL,
total_amount REAL,
FOREIGN KEY(customer_id)REFERENCES customers(customer_id),
FOREIGN KEY(book_id) REFERENCES books(book_id)
)
""")
conn.commit()
```

```python
#diplay the schema of eachntable using pragma
Cursor.execute("PRAGMA table_info(books)")
print(Cursor.fetchall())
Cursor.execute("PRAGMA table_info(customers)")
print(Cursor.fetchall())
Cursor.execute("PRAGMA table_info(orders)")
print(Cursor.fetchall())
```

```python
#Task3
# pandas integration and display first 3 rows
import pandas as pd
books_df=pd.read_sql("SELECT * FROM books",conn)
customers_df=pd.read_sql("SELECT * FROM customers",conn)
orders_df=pd.read_sql("SELECT * FROM orders ",conn)
print(books_df.head(3))
print(customers_df.head(3))
print(orders_df.head(3))
```

```python
#merge data
report = orders_df.merge(customers_df,on="customer_id").merge(books_df,on="book_id")

report[['order_id','name','city','title','quantity','total_amount']]
```

```python
#Average order value
report['total_amount'].mean()
```

```python
#most popular book
report.groupby('title')['quantity'].sum().sort_values(ascending=False)
```

```python
#create data frame
import pandas as pd
discount_data={"book_id":[1,2,3,4,5],'discount_percent':[10,15,5,20,12]}
discount_df=pd.DataFrame(discount_data)
print(discount_df)
```

```python
#save dataframe to sql table
import sqlite3
conn=sqlite3.connect("bookstore.db")
discount_df.to_sql("discounts",conn,if_exists="replace",index=False)
```

```python
# verify the table ceated
query="SELECT * FROM discounts"
pd.read_sql(query,conn)

```

```python
#sql join query(Books+discounts)
query = """
SELECT
b.title,
b.price,
d.discount_percent,
b.price - (b.price * d.discount_percent / 100) AS discounted_price
FROM books b
JOIN discounts d
ON b.book_id = d.book_id
"""

pd.read_sql(query, conn)
```
